**handling missing value**

```python
# np.isnan - detect missing value
# np.nan_to_num() - replace the missin value with different number
# np.isinf() - detect the infinite value

```

detect missing value

```python
import numpy as np
arr = np.array([1,2,np.nan,4])
print(np.isnan(arr))
```

replace nan value


```python
import numpy as np
arr = np.array([1,2,np.nan,4])
cleaned = (np.nan_to_num(arr,nan=100))
print (cleaned)
```

handling infinity value

```python
import numpy as np
arr = np.array([1,2,3,np.nan,np.inf,-np.inf])
cleaned = np.nan_to_num(arr,nan=9, posinf=6,neginf=8)
print(cleaned)

#nan for - null value
#poinf for - positive infinite value
#neginf for - negative infinite value
#for the handling the infinite vlaue syntax - np.nan_to_num(name_the_arr, nan=  )
```
